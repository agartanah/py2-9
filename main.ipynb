```python
from PyPDF2 import PdfReader

reader = PdfReader("texts.pdf")

full_text = ""

for page in reader.pages:
    text = page.extract_text()

    if text:
        full_text += text + "\n"

print(full_text)
```

```python
import re

pattern = r'В\s*а\s*р\s*и\s*а\s*н\s*т[\s\xa0]*\d+[\.\)]?'
raw_docs = re.split(pattern, full_text)

raw_docs = raw_docs[1:]

print(f"Количество документов (вариантов): {len(raw_docs)}")
```

```python
def lemmatize_document(text):
    doc = nt.Doc(text)
    doc.segment(segmenter)
    doc.tag_morph(morph_tagger)

    lemmas = []
    for token in doc.tokens:
        token.lemmatize(morph_vocab)
        if token.lemma and token.lemma.isalpha():
            lemmas.append(token.lemma.lower())

    return lemmas
```

```python
lemmatized_docs = [lemmatize_document(doc) for doc in raw_docs]
```

```python
vocab = set()
for doc in lemmatized_docs:
    vocab.update(doc)

vocab = sorted(vocab)
word_to_index = {word: i for i, word in enumerate(vocab)}

numdocs = len(lemmatized_docs)
numwords = len(vocab)

print("Документов:", numdocs)
print("Слов в словаре:", numwords)
```

```python
import numpy as np

word_frequency = np.zeros(numwords)
for word, idx in word_to_index.items():
    word_frequency[idx] = sum(1 for doc_lemmas in lemmatized_docs if word in doc_lemmas)

idf = np.log(numdocs / word_frequency)
```

```python
tf_vector = np.zeros(numwords)
for word, count in tf.items():
    if word in word_to_index:
        tf_vector[word_to_index[word]] = count
```

```python
tfidf_vector = tf_vector * idf
```

```python
tfidf_dict = {word: tfidf_vector[idx] for word, idx in word_to_index.items() if tfidf_vector[idx] > 0}
print(tfidf_dict)
```

```python
def extract_words(sentence):
    ignore_words = stopwords.words()
    words = re.sub("[^\w]", " ", sentence).split() # заменяет все специальные символы на ' '
    words = [word.lower() for word in words]
    words_cleaned = [w.lower() for w in words if w not in ignore_words]
    return words_cleaned
```

```python
# вычисляет частоту появления слов
def map_book(hash_map, tokens):
    if tokens is not None:
        for word in tokens:
            # слово присутствует
            if word in hash_map:
                hash_map[word] = hash_map[word] + 1
            else:
                hash_map[word] = 1
        return hash_map
    else:
        return None
    
```

```python
def make_hash_map(df):
    hash_map = {}
    for index, row in df.iterrows():
        hash_map = map_book(hash_map, extract_words(row['text']))
    return hash_map
```

```python
def frequent_vocab(word_freq, max_features):
    counter = 0 # инициализирует счетчик значением ноль
    vocab = [] # создает пустой список, который называется vocab
    # перечисляет слова в словаре в порядке убывания частоты
    for key, value in sorted(word_freq.items(), key=lambda item: (item[1], item[0]), reverse=True):
        # функция цикла для получения топ (max_features) количества слов
        if counter < max_features:
            vocab.append(key)
            counter+=1
        else: break
    return vocab
```

```python
hash_map = make_hash_map(df) # создает hash map (слово-частота) из токенизированного набора данных
vocab=frequent_vocab(hash_map, 500)
vocab
```

```python
# Определяем функцию bagofwords со следующими входными данными: sentence и words
def bagofwords(sentence, words):
    sentence_words = extract_words(sentence) #токенизирует предложения/твиты и присваивает их значение переменной sentence_words
    # подсчитывает частоту появления слова
    bag = np.zeros(len(words)) #создает массив NumPy, состоящий из нулей с размером len(words)
    # Циклически перебираем данные и добавляем значение 1, когда токен присутствует в твите
    for sw in sentence_words:
        for i,word in enumerate(words):
            if word == sw:
                bag[i] += 1
    return np.array(bag) # возвращает мешок слов для одного твита
```

```python
# настройте массив NumPy с заданным размером для хранения мешка слов
n_words = len(vocab)
n_docs = len(df)
bag_o = np.zeros([n_docs,n_words])
# используйте цикл, чтобы добавить новую строку для каждого твита
for ii in range(n_docs):
# вызывает предыдущую функцию 'bagofwords'. Обратите внимание на входные данные: sentence и words
    bag_o[ii,:] = bagofwords(df['text'].iloc[ii], vocab)
    bag_o.shape
```

```python
sorted(hash_map.items(), key=lambda item: item[1], reverse=True)[:20]
```

```python
#инициализируйте 2 переменные, представляющие количество твитов (numdocs) и количество токенов/слов (numwords)
numdocs, numwords = np.shape(bag_o)
#Переход к формуле tfidf, как указано выше
N = numdocs
word_frequency = np.empty(numwords)
#Подсчет количества документов, в которых появляется слово
for word in range(numwords): 
    word_frequency[word]=np.sum((bag_o[:,word]>0))
idf = np.log(N / word_frequency)
idf.shape
```

```python
tfidf = np.empty([numdocs, numwords])
```

```python
#циклически перебирает твиты, перемножая частоту появленияслов (представленную мешком слов) с idf
for doc in range(numdocs):
    tfidf[doc, :]=bag_o[doc, :]*idf
    
tfidf.shape
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import copy

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

num_features = X_train.shape[1]  # 500
num_classes = 2

model = nn.Sequential(
    nn.Linear(num_features, 512),
    nn.BatchNorm1d(512),
    nn.ReLU(),
    nn.Dropout(0.5),

    nn.Linear(512, 256),
    nn.BatchNorm1d(256),
    nn.ReLU(),
    nn.Dropout(0.4),

    nn.Linear(256, 128),
    nn.ReLU(),

    nn.Linear(128, num_classes)
).to(device)
```

```python
X_train_t = torch.tensor(X_train, dtype=torch.float32).to(device)
X_test_t  = torch.tensor(X_test,  dtype=torch.float32).to(device)

y_train_t = torch.tensor(y_train, dtype=torch.long).to(device)
y_test_t  = torch.tensor(y_test,  dtype=torch.long).to(device)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)

best_accuracy = 0.0
best_model_state = None
patience = 5
wait = 0
EPOCHS = 50

for epoch in range(EPOCHS):
    # --- TRAIN ---
    model.train()
    optimizer.zero_grad()

    outputs = model(X_train_t)
    loss = criterion(outputs, y_train_t)
    loss.backward()
    optimizer.step()

    # --- EVAL ---
    model.eval()
    with torch.no_grad():
        outputs_test = model(X_test_t)
        _, preds = torch.max(outputs_test, 1)
        acc = (preds == y_test_t).float().mean().item()

    print(
        f"Epoch {epoch+1}/{EPOCHS} | "
        f"Loss: {loss.item():.4f} | "
        f"Acc: {acc:.4f}"
    )

    if acc > best_accuracy:
        best_accuracy = acc
        best_model_state = copy.deepcopy(model.state_dict())
        wait = 0
    else:
        wait += 1
        if wait >= patience:
            print(f"Ранняя остановка на эпохе {epoch+1}")
            break

model.load_state_dict(best_model_state)
print(f"\nЛучшая точность: {best_accuracy * 100:.2f}%")
```
